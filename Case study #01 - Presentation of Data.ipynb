# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(8,6))
sns.countplot(x ='Buyer Gender', data =data )
plt.title('Male VS Female Buyers participated in the sales', fontsize=18)
data['Buyer Gender'].value_counts()
```

# 4. Find the top 5 cars based on their sales price

```python
Top_car = data.sort_values(by='Sale Price', ascending = False)
t=Top_car[['Make','Model','Nickname','Car Gender','Sale Price']]
t.iloc[0:5,:]
```

These are the top 5 cars based on their sales price

# 5. Find the least 5 cars based on their Resell price.

```python
leastcar = data.sort_values(by='Resell Price')
l=leastcar[['Make','Model','Nickname','Car Gender','Resell Price']]
l.iloc[0:5,:]
```
